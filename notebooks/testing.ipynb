```python
import yfinance as yf
import mplfinance as mpf

# Fetch historical data for a stock
ticker = "EURUSD=X"
data = yf.download(ticker, start="2018-01-01", end="2022-12-31")

# Create a candlestick chart with moving averages
mpf.plot(data, 
         type='candle', 
         mav=(50, 100, 200),  # 20 and 50 periods moving averages
         volume=True, 
         title=f"{ticker} Candlestick Chart with MA",
         style='charles')
```
